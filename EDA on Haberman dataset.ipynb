```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
haberman= pd.read_csv(r"C:\Users\KHUSHI\Documents\Python Scripts\haberman.csv")
haberman
```

```python
survival=haberman.loc[haberman["status"]==1];
death=haberman.loc[haberman["status"]==2];
```

```python
sns.FacetGrid(haberman,hue="status",height=5) \
   .map(sns.distplot,"age") \
   .add_legend();
plt.show()
```

```python
sns.FacetGrid(haberman,hue="status",height=5) \
   .map(sns.distplot,"year") \
   .add_legend();
plt.show()
```

```python
sns.FacetGrid(haberman,hue="status",height=5) \
   .map(sns.distplot,"nodes") \
   .add_legend();
plt.show()
```

```python
counts,bin_edges=np.histogram(survival['age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
counts,bin_edges=np.histogram(death['age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

```python
counts,bin_edges=np.histogram(survival['year'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
counts,bin_edges=np.histogram(death['year'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

```python
counts,bin_edges=np.histogram(survival['nodes'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
counts,bin_edges=np.histogram(death['nodes'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
